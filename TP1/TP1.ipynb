>**Question 1 :** 
Commençons par calculer $\mathbb{E}[X_1]$: 
\begin{align*}
    \mathbb{E}_\theta[X_1] &= \frac{1}{\theta}\int_0^\theta xdx \\
                            &= \frac{\theta}{2}
\end{align*}
Pour estimer $\theta$ par la méthode des moments, on doit résoudre l'équation  $\frac{1}{2}\hat \theta_1^n = \frac{1}{n}\sum_{i = 1}^n X_i$ et on obtient donc : 
\begin{equation}
    \hat \theta_1^n = \frac{2}{n} \sum_{i = 1}^n X_i
\end{equation}

> **Question 2:** On utilise la décomposition biais-variance du risque quadratique : 
\begin{equation*}
    L(\theta, \hat \theta_1^n) = \mathbb{E}_\theta[|\theta - \hat \theta_1^n|^2] = (\theta - \mathbb{E}_\theta[ \hat \theta_1^n])^2 + Var_\theta(\hat \theta_1^n)
\end{equation*}
Il est immediat, par linéarité de l'espérence, que le biais de l'estimateur est nul. Il suffit donc de caluler sa variance. Dans le calcul suivant, on utilise le fait que pour $a,b \in \mathbb{R}$ et $X$ une variable aléatoire $Var(aX+b)=a^2 Var(X)$ et que si $Y$ est indépendante de $X$ alors $Var(X+Y) = Var(X)+Var(Y)$.
\begin{align*}
    Var_{\theta}(\hat \theta_1^n) &= Var_{\theta}(\frac{2}{n}\sum_{i = 1}^nX_i) \\
                                &= \frac{4}{n^2}\sum_{i = 1}^n Var_{\theta}(X_i) \\
                                &= \frac{4}{n}Var_{\theta}(X_1)
\end{align*}
Par ailleurs, on a $\mathbb{E}_{\theta}[X_1^2] = \frac{\theta^2}{3}$ et donc 
\begin{equation*}
    Var_{\theta}(X_1) = \mathbb{E}_\theta[X_1^2]-\mathbb{E}_{\theta}[X_1]^2= \frac{\theta^2}{12}
\end{equation*} 
Pour conclure : 
\begin{equation}
    L(\theta, \hat \theta_1^n)= \frac{\theta^2}{3n}
\end{equation}

> **Question 3:** 
Par indépendance, la densité jointe de $(X_1, \dots, X_n)$ est $p_\theta(x_1, \dots, x_n)= \frac{1}{\theta^n}1\!\!1_{[0, \theta]^n}(x_1, \dots, x_n)$. A $x_1, \dots, x_n$ fixé, $\theta \mapsto p_\theta(x_1, \dots, x_n)$ est décroissante sur $[\max(x_1, \dots, x_n), +\infty)$ et s'annule sur $[0, \max(x_1, \dots, x_n))$. On a alors existence et unicité de l'estimateur de maximum de vraissamblance défini de la manière suivante : 
\begin{equation}
    \hat \theta_2^n = \max(X_1, \dots, X_n)
\end{equation}

> **Question 4:** 
Notons $F_\theta^n$ la fonction de répartition de $\hat \theta_2^n$, $F_\theta^n: u \mapsto \mathbb{P}_\theta(\hat \theta_2^n \le u)$. En raisonnant par cas, on obtient l'expression suivante : 
\begin{equation*}
    F_\theta^n(u) =  
    \begin{cases}
        0 \text{ si } u <0\\
         (\frac{u}{\theta})^n \text{ si } 0 \le u \le \theta \\
        1 \text{ sinon}
    \end{cases}
\end{equation*}
Par ailleurs, puisque $X_1, \dots, X_n$ admettent des densités par rapport à la mesure de Lebesgue, par le théorème de Radon-Nikodym $\hat \theta_2^n= \max(X_1, \dots, X_n)$ admet aussi une densité par rapport à la mesure de Lebesgue. En effet, si $A$ est un ensemble Lebesgue négligeable, $\mathbb{P}_\theta(\max(X_1, \dots, X_n) \in A) \le \sum_{i = 1}^n \mathbb{P}_\theta(X_i \in A) = 0$. De plus on peut calculer cette densité de probabilité $f_\theta^n$ en dérivant la fonction de répartition : $f_{\theta}^n : x \mapsto \frac{n}{\theta^n}x^{n-1}1\!\!1_{[0, \theta]}(x)$. On se sert de cette densité pour calculer les moments d'ordre 1 et 2 de $\hat \theta_2^n$ : 
\begin{align*}
    \mathbb{E}_\theta[\hat \theta_2^n] &= \int_0^\theta x f_\theta^n(x)dx \\
                                    &= \frac{n}{\theta^n}\int_0^\theta x^n dx \\
                                    &= \frac{n\theta}{n+1} \\
    \\
    \mathbb{E}_\theta[(\hat \theta_2^n)^2] &= \int_0^\theta x^2 f_\theta^n(x)dx \\
                                        &= \frac{n}{\theta^n}\int_0^\theta x^{n+1}dx \\
                                        &= \frac{n\theta^2}{n+2}
\end{align*}
On peut ensuite calculer le risque quadratique par décomposition biais variance : 
\begin{align*}
    L(\theta, \hat \theta_2^n) &= (\mathbb{E}_\theta[\hat \theta_2^n]- \theta)^2 + Var_\theta(\hat \theta_2^n) \\
                            &= (\mathbb{E}_\theta[\hat \theta_2^n]- \theta)^2 + \mathbb{E}_\theta[(\hat\theta_2^n)^2]- \mathbb{E}_\theta[\hat\theta_2^n]^2 \\
                            &= \Big(\frac{n\theta}{n+1}-\theta\Big)^2+\frac{n\theta^2}{n+2}-\Big(\frac{n\theta}{n+1}\Big)^2 \\
                            &= \frac{(1-n)\theta^2}{n+1} + \frac{n\theta^2}{n+2} \\
                            &= \frac{2\theta^2}{(n+1)(n+2)}
\end{align*}

>**Question 4:** Pour comparer les estimateurs, on compare leur risque associés : 
\begin{align*}
    L(\theta, \hat \theta_1^n) \ge L(\theta, \hat\theta_2^n) &\iff \frac{\theta^2}{3n} \ge \frac{2\theta^2}{(n+1)(n+2)} \\
                                                            &\iff 6n \le (n+1)(n+2) \\
                                                            &\iff n^2-3n + 2 \ge 0 \\
\end{align*}
Cette condition est toujours vérifiée avec égalité pour $n\in\{1,2\}$. Ainsi pour un ou deux échantillons, les deux estimateurs sont de même qualité, mais sinon il est toujours préférable de choisir l'estimateur de maximum de vraissamblance.

>**Question 2:** 
L'estimateur $\hat \omega$ de $\bar \omega^*$ proposé par le modèle Adaline est solution du problème d'optimisation suivant : 
\begin{equation}
     \hat \omega \in \argmin_{\bar \omega \in \mathbb{R}^3 } \mathbb{E}_{(x,y)\sim \mathbb{P}}[(y-\langle\bar x, \bar \omega\rangle)^2]
\end{equation}
Il s'agit exactement de la forme de problème que l'algorithme du gradient stochastique (SGD) permet de résoudre. On défini alors  la fonction $j:(\bar \omega,x,y) \mapsto (y-\langle\bar x, \bar \omega\rangle)^2$ qui a pour gradient $\nabla_{\bar\omega}j(\bar \omega,x,y)= -2(y-\langle \bar x, \bar \omega \rangle)\bar x $.
L'algorithme SGD, dans le cadre de l'apprentissage supervisé (où l'on échantillonne pas une distribution mais on dispose de données), se formule en alternant les instructions suivantes après initialisation $\hat \omega_0$ : 
\begin{align*}
&\text{1. Selectionner $(x_{k+1},y_{k+1})$ de manière aléatoire et uniforme dans le jeu de données} \\
&\text{2. } \hat \omega_{k+1} \gets \hat \omega_k - \varepsilon_k \nabla_{\bar \omega}j(\hat \omega_k, x_{k+1}, y_{k+1})
\end{align*}
où $(\varepsilon_k)_{k \ge 0}$ est une suite positive vérifiant $\sum_{k =1}^\infty \varepsilon_k = \infty$ et $\sum_{k =1}^\infty \varepsilon_k^2 < \infty$, en particulier on peut choisir $\varepsilon_k = \frac{1}{(k+1)^\alpha}$ où $\alpha \in ]\frac{1}{2},1]$. On choisit d'arrêter la descente de gradient au bout d'un certain nombre d'itérations fixé. On se questionne aussi sur la validité des hypothèses du théorème de convergence de la SGD, en reprenant les notations du cours : (H1) aprori on ne connait pas la distribution des données mais pour que le problème soit bien posé on peut supposer que pour tout $\bar\omega \in \mathbb{R}^3$, $j(\bar \omega, x,y)$ admet une espérence ; (H2) pour tout $(x,y)\in \mathbb{R}^2 \times \{-1,1\}$,  $\bar \omega  \mapsto j(\bar \omega, x, y)$ est la composition d'une fonction affine et convexe et est donc convexe, de plus c'est aussi une fonction $C^1$ ; (H3) cette hypothèse n'est pas vérifiée car pour $(x,y)$ la fonction $\bar \omega \mapsto \nabla_{\bar \omega} j(\bar \omega, x,y)$ est affine non nulle et n'est donc pas bornée ; (H4) sans connaissance à priori sur la distribution des données, il n'est pas possible d'affirmer la validité de cette hypothèse ; (H5) est vérifiée par construction des pas d'apprentissage.



>**Question 3:**

```python
def sgd(alpha, n_iter, x_bar, y, key, target = None):
    """
    Implementation de l'algorithme SGD pour l'apprentissage d'un modèle Adaline.

    Args:
        alpha: float, exposant du taux d'apprentissage epsilon_k = 1/(k+1)^alpha
        n_iter: int, nombre d'itérations
        x_bar: jnp.ndarray, shape (N, d), données d'entrée
        y: jnp.ndarray, shape (N,), labels cibles
        key: jax.random.PRNGKey, clé pour le générateur de nombres aléatoires
        target: np.ndarray, shape (d,), cible de l'apprentissage, si None, on ne calcule pas le loss
    
    Returns:
        omega: jnp.ndarray, shape (d,), estimation de la cible
        dist_to_target: jnp.ndarray, shape (n_iter), évolution distance à la cible si target est donné, 0 sinon
    """
    def step(carry, sample_idx):
        k, omega = carry["k"], carry["omega"]
        xk_bar, yk = x_bar[sample_idx], y[sample_idx]
        epsilon_k = 1/(k+1)**alpha
        omega_new = omega +  epsilon_k * (yk - xk_bar @ omega) * xk_bar
        if target is not None:
            dist_to_target = jnp.linalg.norm(omega_new - target)
        else:
            dist_to_target = 0
        return {"k": k+1, "omega": omega_new}, dist_to_target
    dim = x_bar.shape[1]
    N_samples = x_bar.shape[0]
    subkey1,subkey2 = jr.split(key)
    w_init = jr.normal(subkey1, (dim,)) # Initialize at random
    init = {"k": 0, "omega": w_init}
    samples_idx = jr.randint(subkey2, (n_iter,), 0, N_samples) # Random sample per iteration
    results, dist_to_target = jax.lax.scan(step, init, samples_idx)
    return results["omega"], dist_to_target
    
```

> On observe 10 trajectoires de l'algorithme avec $\alpha = 1$. Aprèsde 500 itérations, l'algorithme semble atteindre un régime stationnaire. En revanche, les trajectoires ne semblent pas converger vers l'objectif $\omega^*$ et la distance entre $\hat \omega$ et $\omega^*$ semble stationner à $0.7$. Cela peut être due à la convergence $L^2$, et non presque-sûre, de l'algorithme ou bien aux hypothèses (H3) et (H4) du théorème qui ne sont pas vérifiées.

```python
key, subkey = jr.split(key, 2)
subkeys = jr.split(subkey, 10)
trajectories = jax.vmap(lambda subkey: sgd(1, 500, x_bar, y, subkey, target = w_star)[1])(subkeys)
```

```python
fig, ax = plt.subplots()
for i in range(10):
    ax.plot(trajectories[i,:])
ax.set_title('Convergence for 10 different trajectories')
ax.set_xlabel('Iteration')
ax.set_ylabel('Distance to target')
ax.set_yscale('log')
```

> On s'intéresse dans le tracé suivant à l'influence de $\alpha$ sur la qualité de l'estimateur obtenu. Pour cela on trace la valeur $d(\hat \omega, \omega^*)$ moyennée sur 1000 trajectoires en fonction de la du $\alpha$ utilisé pour faire l'esimation. La courbe obtenue indique que la qualité de l'estimation augmente lorsque $\alpha$ se rapproche de $1$.

```python
N_avg = 10000
n_iter = 500
alphas = jnp.linspace(0.5 + 1e-6, 1, 100)
key, subkey = jr.split(key, 2)
subkeys = jr.split(subkey, 100)

def avg_dist_to_target(alpha, n_iter, n_avg, x_bar, key): 
    subkeys = jr.split(key, n_avg)
    dists_to_target = jax.vmap(lambda subkey: sgd(alpha, n_iter, x_bar, y, subkey, target = w_star)[1][-1])(subkeys)
    return jnp.mean(dists_to_target)

avg_distances = jax.vmap(lambda alpha, key: avg_dist_to_target(alpha, n_iter, N_avg, x_bar, key))(alphas, subkeys)
```

```python
fig, ax = plt.subplots()
ax.plot(alphas, avg_distances)
ax.set_title('Average distance to target for different values of alpha')
ax.set_xlabel('Alpha')
ax.set_ylabel('Average distance to target')
```

Pour les labels $y$, on s'intéresse uniquement à la donnée binaire malade/non malade, les labels $0$ sont donc remplacés par $-1$ et $1,2,3,4$ par $1$ . On divise aussi notre dataset en une section d'entrainement et une section de test.

```python
X_clean = X.dropna()
y_clean = y.loc[X_clean.index]
x = jnp.array(X_clean)
y = jnp.where(y_clean.values == 0, -1.0, 1.0)
scaler = StandardScaler()
x = scaler.fit_transform(x)
x_bar = jnp.column_stack([x, jnp.ones(x.shape[0])])
y = y.reshape(len(y))
```

```python
train_perc = 0.8 # Proportion of the dataset used for training

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = train_perc, random_state = 0, stratify = y)
x_bar_train = jnp.column_stack([x_train, jnp.ones(x_train.shape[0])])
x_bar_test = jnp.column_stack([x_test, jnp.ones(x_test.shape[0])])
```

>On évalue maintenant le modèle sur le jeu de données en en testant différentes valeurs du paramètre $\alpha$. Le tracé obtenu semble indiquer une valeur optimale de $\alpha$ à environ $0.9$. Pour un tel $\alpha$ on obtient une précision de $0.86$, ce qui est très satisfaisant en comparaison des performances de références présentées dans la documentation de la base de données. 

```python
def accuracy_model(alpha, key):
    # Train the model
    w_hat, _ = sgd(alpha, 50000, x_bar_train, y_train, key)
    # Predict on the test set
    preds = jnp.where((x_bar_test @ w_hat) >= 0, 1, -1)
    # Compute accuracy as fraction of correct predictions (returns value in [0,1])
    return jnp.mean(preds == y_test)
```

```python
N_avg = 100
key, subkey = jr.split(key)
alphas = jnp.linspace(0.5 + 1e-2, 1, 100)
subkeys_2d = jr.split(subkey, N_avg*len(alphas)).reshape(len(alphas), N_avg, -1)

# Computes mean accuracy over N_avg trials for each alpha
mean_accuracies = jax.vmap(
    lambda alpha, subkeys: jax.vmap(
        lambda key: accuracy_model(alpha, key)
    )(subkeys).mean()
)(alphas, subkeys_2d)
```

```python
fig, ax = plt.subplots()    
ax.plot(alphas, mean_accuracies)
ax.set_title(f'Accuracy of the model averaged over {N_avg} trials')
ax.set_xlabel('Alpha')
ax.set_ylabel('Accuracy')
plt.show()
```
